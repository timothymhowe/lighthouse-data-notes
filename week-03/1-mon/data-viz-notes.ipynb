## Getting Comfortable with ```matplotlib```

#### A functional approach

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


x = np.linspace(0,10,20)
y = x**2


````python
plt.plot(x,y)
````

````python
plt.title("Hello World!")
plt.xlabel("My X axis")
plt.ylabel("My Y Axis")
````

````python
plt.subplot(1,2,1)
plt.plot(x,y,'red')

plt.subplot(1,2,2)
plt.plot(y,x,'green')
````

#### An object oriented approach! (preferred method)

````python
fig = plt.figure()
ax = fig.add_axes([0.1,0.2,0.8,0.9])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('The title')

ax.plot(x,y,'purple')
````

````python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
````

````python
fig = plt.figure(figsize = (8,2), dpi = 100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
````

````python
fig, ax = plt.subplots(nrows=2,ncols=1,figsize=(8,4), dpi=100)
ax[0].plot(x,y)
ax[1].plot(y,x)

plt.tight_layout()
````

### Legends

````python
fig = plt.figure(figsize=(8,6),dpi = 60)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = "X^2 Plot")
ax.plot(x,x**3, label = "X^3 Plot")

ax.legend()
````

## Matplotlib Walkthrough

````python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,6)
````

````python
# Line Plot
x = np.linspace(0,100,200)
y = np.sin(x)

plt.plot(x,y,'red');
````

````python
# Scatter Plot
x = 10*np.random.rand(200,1)
y = 15*np.random.rand(200,1)

plt.scatter(x,y);
````

````python
# Histogram
y = 15*np.random.rand(200,1);
plt.hist(y,bins=20);
````

## Graphs on Common Axes:

````python
x = np.linspace(0,100,200)
y1 = np.sin(x)
y2 = np.sin(x) * np.cos(x)

plt.plot(x,y1,'red')
plt.plot(x,y2,'blue');
````

## Legends and Annotations:


````python
import pandas as pd
from sklearn import datasets
````

````python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_name'] = iris_df.target.map(
    {0: 'setosa',
    1: 'versicolor',
    2: 'virginica'}
)
````

````python
iris_df.head()
````

````python
setosa = iris_df[iris_df['target_name'] == 'setosa']

versicolor = iris_df[iris_df['target_name'] == 'versicolor']

virginica = iris_df[iris_df['target_name'] == 'virginica']

plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'],
           marker='o',color='red',label='setosa')


plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'],
           marker='o',color='teal',label='versicolor')


plt.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'],
           marker='o',color='pink',label='virginica')


plt.annotate('setosa', xy = (5.0,3.5), xytext = (4.25,4.0), arrowprops={'color':'red'})
plt.annotate('versicolor', xy = (7.2,3.6), xytext=(6.5,4.0),arrowprops={'color':'red'})
plt.annotate('virginica', xy=(5.05,1.95), xytext=(5.5,1.75),arrowprops={'color':'red'})



plt.legend(loc='upper right')
plt.title('Iris flower')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
````
