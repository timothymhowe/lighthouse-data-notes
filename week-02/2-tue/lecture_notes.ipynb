```python
import pandas as pd
```

```python
pokemon = pd.read_csv('data/pokemon.csv')
```

```python
pokemon.describe()
```

```python
pokemon_2 = pokemon.copy()
pokemon_2['Total Atk'] =  pokemon_2['Attack'] + pokemon_2['Sp. Atk']
```

```python
pokemon_2.groupby('Legendary').mean()
```

```python
pokemon_2.sort_values(by = ['Attack','HP'],ascending = [False,True])
```

```python
the_filter = pokemon['Total'] >= 500
```

```python
pokemon.where(the_filter,'Weak Pokemon')
```

```python
pokemon[the_filter]
```

```python
pokemon_aggregator = pokemon.copy()
```

```python
pokemon_aggregator.groupby(['Type 1','Type 2']).mean()
```

```python
pokemon_aggregator.groupby(['Type 1']).mean().sort_values(by=['Total'],ascending = False)
```

```python
pokemon_aggregator.groupby(['Type 1']).agg(['max','min','mean']).sort_values(by=[('Total','mean')],ascending = False)

pokemon_agg_dict = {
    '#':[],
    'Total':['max','mean','min']
}
```
